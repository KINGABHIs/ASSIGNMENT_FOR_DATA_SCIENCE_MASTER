## Q1. Explain GET and POST methods.

### Ans:
> GET Method
* Purpose: Used to request data from a server.
* Data Sending: Sends data in the URL as query parameters (visible to users).
* Use Case: When no sensitive data is involved, like search queries.
* Caching: Data can be cached.
* Limit: Limited data size due to URL length.


> POST Method
* Purpose: Used to send data to a server to create/update resources.
* Data Sending: Sends data in the body of the request (not visible in URL).
* Use Case: For sensitive data like passwords or large data submissions (e.g., forms).
* Caching: Not cached.
* Limit: No size limit in the body.

 GET is for fetching data, and POST is for sending data securely to the server.

## Q2. Why is request used in Flask?

### Ans:
In Flask, the request object is used to access data sent by the client to the server.

Key Uses:
1. Get Input Data: Retrieve form data, query parameters, JSON, etc.
> Example: request.form, request.args, request.json.

2. Check Request Type: Identify if the request is GET, POST, etc.
> Example: request.method.

3. Access Headers: Get metadata like content type or authentication tokens.
> Example: request.headers.

request helps handle input and manage client-server communication in Flask.

## Q3. Why is redirect() used in Flask?

### Ans:
In Flask, redirect() is used to send a user to a different URL or page.

Key Uses:
* Page Navigation: Redirect users after a form submission or login.
* Error Handling: Send users to a custom error page.
* Dynamic Routing: Redirect to another route dynamically.

## Q4. What are templates in Flask? Why is the render_template() function used?

### Ans:
Templates in Flask

* Definition: Templates are HTML files used to create dynamic web pages.
* Purpose: Combine static content (HTML) with dynamic data (variables) sent from Flask.
* Example: Displaying user data, search results, or any content dynamically.

use of render_template() 

* Purpose: To render (display) an HTML template with dynamic data.
* How It Works: Passes variables from Flask to the template.
