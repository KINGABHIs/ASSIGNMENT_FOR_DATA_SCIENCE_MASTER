## Q1. What is Flask Framework? What are the advantages of Flask Framework?

### Ans :
Flask : Flask is a lightweight web framework in Python used for building web applications. It is based on the Werkzeug toolkit and Jinja2 template engine. Flask is a "micro-framework," meaning it provides only the essential tools for web development, making it simple and flexible.

* Advantages of Flask Framework

Lightweight: Flask is minimal and doesn't include unnecessary components, so you can add features as needed.

Flexibility: It allows developers to customize applications easily.

Easy to Learn:Ideal for beginners due to its simplicity and clear documentation.

Extensible: Supports extensions for adding features like authentication, databases, etc.

Integrated Tools: Built-in support for routing, templates (Jinja2), and development server.

Scalable: Suitable for small projects and scalable for larger applications.

Community Support: Strong community with plenty of tutorials and resources.

Used for : Building APIs, microservices, and small to medium-sized web applications.










## Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.

### Ans :


```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return ' Hello World!'

if __name__ == '__main__':
    app.run(host = "0.0.0.0")
```

## Q3. What is App routing in Flask? Why do we use app routes?

### Ans :
App Routing in Flask : App routing is the process of defining URLs (routes) in a Flask application that trigger specific functions when accessed. It helps map a URL to a specific Python function (called a view function) that returns a response.

In Flask, the @app.route() decorator is used to define routes.

App Routes Used for :

* URL Mapping: Allows users to access specific pages or actions via meaningful URLs, e.g., /home, /about.

* Organized Code: Keeps the application modular by associating different URLs with different functions.

* Dynamic URLs: Supports dynamic routing, enabling the use of variables in the URL (e.g., /user/<username>).

* User Interaction: Routes are essential to handle user requests and return appropriate responses.

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

### Ans:
The url_for() function is used in Flask for URL building. It generates the URL for a specific function based on its name and any arguments provided.

Used url_for():
* Dynamically build URLs without hardcoding them.
* Avoid issues when routes or query parameters change.
* Simplifies navigation and redirection.
